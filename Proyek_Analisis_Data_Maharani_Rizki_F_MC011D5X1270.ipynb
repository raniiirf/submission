## Analisis Lanjutan (RFM)

```python
rfm_data = all_data.copy()
rfm_data.head()
```

```python
rfm_data.info()
```

```python
now = rfm_data['order_purchase_timestamp'].max()
rfm_recency = rfm_data.groupby('customer_city')['order_purchase_timestamp'].max().reset_index()
rfm_recency['Recency'] = (now - rfm_recency['order_purchase_timestamp']).dt.days
rfm_recency
```

```python
rfm_frequency = rfm_data.groupby('customer_city')['order_id'].count().reset_index(name='Frequency')
rfm_frequency
```

```python
# Calculate Monetary Value
rfm_monetary = rfm_data.groupby('customer_city')['price'].sum().reset_index(name='Monetary')
rfm_monetary
```

```python
# Merge RFM metrics into a single dataframe
rfm_data = pd.merge(rfm_recency, rfm_frequency, on='customer_city', how='inner')
rfm_data = pd.merge(rfm_data, rfm_monetary, on='customer_city', how='inner')
rfm_data
```

```python
# RFM Scoring
rfm_data['R'] = pd.qcut(rfm_data['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_data['F'] = pd.qcut(rfm_data['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm_data['M'] = pd.qcut(rfm_data['Monetary'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])

# Combine RFM Scores
rfm_data['RFM Score'] = rfm_data['R'].astype(str) + rfm_data['F'].astype(str) + rfm_data['M'].astype(str)
rfm_data
```

```python
# Analyze RFM segments
# Example: Group by RFM Score and analyze
rfm_segments = rfm_data.groupby('RFM Score')['customer_city'].count().reset_index(name='Customer Count')

# Print the RFM segments
print(rfm_segments)
```

```python
# Define Segments
def assign_segment(rfm_score):
    rfm_dict = {
        "Champions": [555, 554, 544, 545, 454, 455, 445],
        "Loyal": [543, 444, 435, 355, 354, 345, 344, 335],
        "Potential Loyalist": [553, 551, 552, 541, 542, 533, 532, 531, 452, 451, 442, 441, 431, 453, 433, 432, 423, 353, 352, 351, 342, 341, 333, 323],
        "New Customers": [512, 511, 422, 421, 412, 411, 311],
        "Promising": [525, 524, 523, 522, 521, 515, 514, 513, 425,424, 413,414,415, 315, 314, 313],
        "Need Attention": [535, 534, 443, 434, 343, 334, 325, 324],
        "About To Sleep": [331, 321, 312, 221, 213, 231, 241, 251],
        "At Risk": [255, 254, 245, 244, 253, 252, 243, 242, 235, 234, 225, 224, 153, 152, 145, 143, 142, 135, 134, 133, 125, 124],
        "Cannot Lose Them": [155, 154, 144, 214,215,115, 114, 113],
        "Hibernating customers": [332, 322, 233, 232, 223, 222, 132, 123, 122, 212, 211],
        "Lost customers": [111, 112, 121, 131,141,151]
    }
    for segment, scores in rfm_dict.items():
        if int(rfm_score) in scores:
            return segment
    return 'Others'

rfm_data['Segment'] = rfm_data['RFM Score'].apply(assign_segment)
rfm_data
```

```python
# Count cities per segment
segment_counts = rfm_data['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'City Count']
segment_counts
```

```python
# Visualize Segment Counts
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment', y='City Count', data=segment_counts, palette='viridis')
plt.title('Number of Cities per Segment')
plt.xticks(rotation=45)
plt.show()
```
